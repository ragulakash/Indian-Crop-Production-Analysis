```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/archive (4).zip")
```

```python
print(df.head())
```

```python
df.head()
```

```python
df.describe()
```

```python
print(df.isnull().sum())
```

```python
mean_production = df['Production'].mean()
df['Production'] = df['Production'].fillna(mean_production)

print(df.isnull().sum())
```

```python
#Indian States with the most Crop Production

most_cp = df.groupby('State_Name')['Production'].count().sort_values(ascending=False).head(30)
print(most_cp)
```

```python
#REFER the above code for reference

values = np.random.randint(0, 100, 30)

plt.bar(range(len(most_cp)), most_cp)

plt.xlabel("States")
plt.ylabel("Crop Production")
plt.title("Crop Production")

plt.show()
```

```python
popular_sports = df['State_Name'].value_counts().sort_values(ascending=False).head(30)
popular_sports.plot(kind='bar')
plt.xlabel('States')
plt.ylabel('Production')
plt.show()
```
